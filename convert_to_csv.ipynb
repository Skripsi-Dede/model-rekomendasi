```python
import os
import csv
import shutil

dir_path = "./dataset_resized/"
list_dir = os.path
root_dirs = []
male_dirs = []
female_dirs = []
count = 0

folder_tree = {}

def insert_files_to_dict(sub_data: dict, files: list, key: str):
    sub_data[key] = files

def insert_folder_to_dict(sub_data: dict, key: str):
    sub_data[key] = {}
    return sub_data

folders = []
files_data = []
for root, dirs, files in os.walk(dir_path):
    folder = root.split("/")[2]
    # skip empty folder
    if len(folder) == 0:
        continue
    # split sub folder
    folder = folder.split("\\")
    if len(files) == 0:
        continue
    files_data.append(files)
    if len(folder) == 1:
        folder.insert(0,"all")
    folders.append(folder)

header = ["file_name","gender","category"]

cwd = os.getcwd()
path_to_csv_dir = os.path.join(cwd,"../deploy/pkg/config")
path_to_csv = os.path.join(os.path.join(cwd,"../deploy/pkg/config"),"data.csv")

listdirs = os.listdir(path_to_csv_dir)
if "data.csv" in listdirs:
    print("File found")
    try:
        os.remove(path_to_csv)
    except Exception as e:
        shutil.rmtree(path_to_csv)


file = open(path_to_csv,"a+")
file.close()

with open(path_to_csv, 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    # write the header
    writer.writerow(header)

    for index, item in enumerate(folders):
        for index,file in enumerate(files_data[index]):
            dataset = [file]
            for it in item:
                dataset.append(it)
            writer.writerow(dataset)
            

    # write the data
```
